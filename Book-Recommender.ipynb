```python
#Importing Depndencies
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
#Importing DataSets as CSV Files
books = pd.read_csv("books.csv")
data = pd.read_csv("book_data2.csv")
```

```python
books.head(2)
```

```python
data.head(2)
```

```python
#Merging both dataset to one
book= books.merge(data,on='book_id')
```

```python
book.head()
```

```python
#Extracting columns which is used for future.
book = book[['book_id','book_title','authors','book_desc','genres','image_url_y']]
```

```python
book.head(2)
```

```python
#Checking Null Values.
book.isnull().sum()
#Unfortunately, there are many null values seems in dataset.
```

```python
#Dropping null values
book.dropna(inplace=True)
```

```python
#Checking for duplicates
book.duplicated().sum()
```

```python
#Splitting description and genres section as a list.
book['book_desc']= book['book_desc'].apply(lambda x:x.split())
```

```python
book['genres']= book['genres'].apply(lambda x:x.split("|"))
```

```python
book.head(1)
```

```python
#Making all columns like description and genres columns i.e. in list form.
book['genres']= book['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
book['authors']= book['authors'].apply(lambda x:x.split(","))
```

```python
book['authors']= book['authors'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
book.head(2)
```

```python
#Now, generating tags from those datas
book['tags'] = book['authors'] + book['genres'] + book['book_desc']
```

```python
new_df = book[['book_id', 'book_title', 'tags', 'image_url_y']]
```

```python
#After making tags the finat dataset is ready.
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
#Now import nltk for stemming the similar tags
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#Making a function for stemming
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#Using Bag of Words from that i use count vectorizer to make those tags in vector forms.
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector
```

```python
#values after stemming
cv.get_feature_names()
```

```python
#Putting cosine similarity on those vectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6]
```

```python
#Importing dependencies for fetching images.
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Making function for getting images from given url which are in dataset
#Use of those vectors and countvectorisation recommending books.
def recommend(book):
    book_index = new_df[new_df['book_title']==book].index[0]
    distances = similarity[book_index]
    book_list=sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]
    book_indices=[i[0] for i in book_list]
    rec = new_df[['book_title','image_url_y']].iloc[book_indices]
    for i in book_list:
        print(new_df.iloc[i[0]].book_title)
  
    for i in rec['image_url_y']:
        response = requests.get(i)
        img = Image.open(BytesIO(response.content))
        plt.figure()
        plt.imshow(img)
```

```python
#Result of recommendation
recommend('The Hunger Games')
```
